## Introduction

The primary aim of this study is to establish a rigorous, evidence-based diagnostic cut-off score for the Shona Symptom Questionnaire (SSQ), specifically the theoretically derived 10-item subset (SSQ-10), using the Patient Health Questionnaire-9 (PHQ-9) as a reference standard for probable depression. To achieve this aim, the study first evaluates the structural validity and stability of the SSQ by comparing the original 14-item scale against the theoretical SSQ-10 and the existing 8-item Dixon-Chibanda subset ([Chibanda et al. 2019](#ref-chibanda2019ssq8)) to identify the optimal configuration for clinical use. Subsequently, the analysis validates measurement invariance to confirm that the SSQ-10 functions consistently across depression status, biological sex, and age groups, ensuring the derived cut-off is universally applicable without bias ([Putnick and Bornstein 2016](#ref-putnick2016measurement)). Finally, the study derives a harmonized diagnostic threshold to facilitate accurate depression screening and comparability between the SSQ-10 and the PHQ-9 among adolescents and young adults.

## Methods

### Setting and Study Design

This study is embedded within the Isisekelo Sempilo (“Foundation of Life”) trial, a $2 \times 2$ factorial randomized controlled trial conducted by the Africa Health Research Institute (AHRI) in the uMkhanyakude district of rural KwaZulu-Natal, South Africa ([Chidumwa et al. 2022](#ref-chidumwa2022isisekelo)). The trial aimed to evaluate interventions integrating HIV prevention and care within sexual and reproductive health services. The study setting is predominantly rural and characterized by high levels of poverty and unemployment, with over 85% of young adults aged 20–24 reported as unemployed. The AHRI demographic surveillance area covers a population of approximately 140,000, characterized by a high prevalence of HIV.

### Participants and Sampling

The trial utilized the AHRI demographic surveillance system as a sampling frame to randomly select 3,000 men and women aged 16–29 years, stratified by sex. Eligibility criteria for the main trial included residence within the surveillance area, willingness to provide informed consent, and willingness to participate in follow-up assessments including HIV testing via dried blood spot.

For the present psychometric analysis, a specific subsample was selected based on age, restricting the cohort to participants aged 16–24 years, and data completeness, requiring concurrent baseline measurements for both the SSQ-14 and the PHQ-9. This selection ensures that the validation and harmonization analyses are conducted on a directly comparable cohort relevant to adolescent and young adult mental health.

### Data Collection Procedures

Participants were invited to participate via home visits conducted by trained fieldworkers. Upon contact, the study purpose was explained, and informed consent was obtained (written consent for adults; assent plus parental/guardian consent for participants aged 16–17).

Participants completed a structured baseline questionnaire administered electronically by fieldworkers using secure tablets. The interview included sociodemographic details, health care utilization history, and the full SSQ-14 and PHQ-9 instruments. To minimize social desirability bias and ensure comprehension, interviews were conducted in the local language, isiZulu, in private settings within the participant’s household or a preferred community location. Clinical attendance data were captured electronically via the ClinicLink system across 11 clinics within the surveillance area.

### Measures

**The Patient Health Questionnaire (PHQ-9)**

The PHQ-9 is the depression module of the PRIME-MD diagnostic instrument, designed to align strictly with the criteria for Major Depressive Disorder (MDD) in the DSM-IV ([Kroenke, Spitzer, and Williams 2001](#ref-Kroenke2001)). It consists of 9 items scored on a 4-point Likert scale ranging from “Not at all” (0) to “Nearly every day” (3). It measures the severity of MDD symptoms over the past 2 weeks, covering domains such as anhedonia, depressed mood, sleep disturbance, fatigue, and suicidal ideation. While validated in South Africa, its etic approach may miss non-DSM manifestations of distress common in African contexts.

**The Shona Symptom Questionnaire (SSQ-14)**

The SSQ-14 represents an emic alternative. Developed in Zimbabwe, it was derived from ethnographic research into how Shona speakers articulate mental distress ([Patel et al. 1997](#ref-Patel1997)). It consists of 14 items with a binary (Yes/No) response format and measures Common Mental Disorders (CMD)—a blend of depression and anxiety—over the past 7 days. It includes indigenous idioms such as kufungisisa (“thinking too much”) and somatic sensations. While culturally resonant, its binary format and mixing of anxiety and depression items present specific challenges for harmonisation with the polytomous, depression-specific PHQ-9 ([Haney et al. 2014](#ref-Haney2014)).

**Derivation of the SSQ-10 (Theoretical Subset)**

To enhance the specificity of the SSQ for depression screening, a theoretically derived subset (SSQ-10) was configured by retaining items that conceptually align with DSM-5 and ICD-11 criteria for a depressive episode, while excluding non-specific distress items ([Patel et al. 1997](#ref-Patel1997)).

Ten items were retained: thinking too much (cognitive distress/rumination), difficulty concentrating (cognitive), irritability (affective-arousal), poor sleep (somatic-affective), crying (affective), feeling tired (somatic-affective), suicidal thoughts (cognitive-affective), general unhappiness (affective-cognitive), work lagging behind (functional/cognitive), and difficulty deciding (cognitive). Irritability was specifically retained as it is recognized as an alternative mood expression for youth depression in DSM-5.

Four items were excluded from the theoretical subset due to conceptual divergence from the core depression construct. Nightmares (SSQ04) were removed as they overlap with sleep disturbance but align more closely with anxiety or trauma. Seeing or hearing things (SSQ05) was excluded as it represents perceptual disturbances associated with severe or psychotic depression rather than core depressive criteria. Stomach ache (SSQ06) was excluded to reduce confounding with general somatic complaints common in this setting, while being frightened by trivial things (SSQ07) was removed as it serves as a marker for anxiety or hyperarousal rather than depression.

### Data Analysis

All statistical analyses were conducted using R Statistical Software ([R Core Team 2023](#ref-rcoreteam2023)).

**Descriptive Statistics**

Descriptive statistics were computed to summarize the demographic and clinical characteristics of the sample. Categorical variables (e.g., biological sex, employment status, depression status) were reported as frequencies and percentages. Continuous variables (e.g., age, PHQ-9 total score) were summarized using means and standard deviations (SD). The distribution of item responses was examined to identify endorsement rates and potential floor or ceiling effects.

**Structural Validity and Stability**

To determine the most robust version of the SSQ for screening purposes, three item configurations were evaluated: the original 14-item scale (SSQ-14), a theoretically derived 10-item subset focusing on core depressive symptoms (SSQ-10), and the 8-item Dixon-Chibanda subset ([Chibanda et al. 2019](#ref-chibanda2019ssq8)).

A Monte Carlo split-sample resampling approach was employed to rigorously test the stability of these structures. The dataset was randomly partitioned into exploratory and confirmatory halves for 10,000 iterations. In each iteration, Exploratory Factor Analysis (EFA) was conducted to identify the most likely factor structure using established criteria appropriate for categorical data, specifically the Kaiser Criterion and Parallel Analysis, utilizing polychoric correlations ([Holgado–Tello et al. 2010](#ref-holgado2010polychoric)). The Consensus Model was then identified as the specific grouping of items that appeared most frequently across these replications, serving as the definitive structure for each scale version.

Following the exploratory phase, the stability of the Consensus Model for the SSQ-10 was tested using Confirmatory Factor Analysis (CFA) on the hold-out data samples using the Weighted Least Squares Mean and Variance adjusted (WLSMV) estimator.

**Reliability and Measurement Invariance**

Internal consistency reliability was calculated using both Cronbach’s Alpha and McDonald’s Omega Total to provide a comprehensive assessment of the scale’s precision.

To ensure the scale performs fairly across different populations, measurement invariance was assessed. This procedure evaluated whether the SSQ-10 measures depression consistently regardless of a participant’s biological sex, age group, or depression severity status ([Putnick and Bornstein 2016](#ref-putnick2016measurement)).

**Diagnostic Cut-offs and Harmonisation**

To drive diagnostic cut-offs and harmonisation, an optimal screening cut-off for the SSQ-10 was established using the PHQ-9 as the clinical reference standard. Two complementary methods were applied. First, Receiver Operating Characteristic (ROC) analysis was used to identify the score threshold that maximized the balance between sensitivity and specificity. Second, Item Response Theory (IRT) was employed to overcome the limitations of simple binary cut-offs by linking the SSQ-10 and PHQ-9 onto a common latent metric of depression severity ([Samejima 1969](#ref-samejima1969estimation)). This allowed for the identification of the precise level of underlying depression corresponding to the PHQ-9 threshold and mapping it directly to the SSQ-10, creating a harmonized standard for classification ([Kolen and Brennan 2014](#ref-kolen2014test)).

## Results

### Participant Characteristics

The demographic and clinical characteristics of the study sample ($N=843$) are summarized in <a href="#tbl-demographics" class="quarto-xref">Table 1</a>. The cohort included a relatively balanced distribution of biological sex, with 51% males and 49% females. The age distribution was skewed slightly towards young adults, with 57% of participants aged 20–24 years and 43% aged 17–19 years. The majority of participants resided in rural households (68%), while 28% lived in peri-urban areas and 3.1% in urban settings.

Regarding socioeconomic status, 44% of participants were currently enrolled in school, and 44% were not employed. Notably, 23% were classified as students and thus not applicable for standard unemployment definitions. Food insecurity was reported by 27% of the sample, and 21% were recipients of a government grant. In terms of health and risk factors, 11% of participants tested positive for HIV, 20% reported experiencing violence, and 52% reported alcohol use. Based on the PHQ-9 reference standard, the prevalence of probable depression (score $\ge$ 10) in the sample was 4.0% ($n=34$).

Item-level analysis of the SSQ-14 revealed significant variability in symptom endorsement between depressed and non-depressed groups (<a href="#tbl-item-dist" class="quarto-xref">Table 2</a>). Participants classified as depressed by the PHQ-9 endorsed SSQ items at substantially higher rates than non-depressed participants. For example, thinking too much (SSQ01) was endorsed by 74% of the depressed group compared to 26% of the non-depressed group, and general unhappiness (SSQ12) was endorsed by 59% versus 11%.

Point-biserial correlations confirmed strong associations between individual SSQ items and the continuous PHQ-9 total score, with coefficients ranging from 0.190 to 0.405 (all $p < 0.001$). The strongest correlations with the PHQ-9 total score were observed for thinking too much ($r_{pb}=0.405$), generally unhappy ($r_{pb}=0.405$), and difficulty concentrating ($r_{pb}=0.400$). Conversely, items excluded from the theoretical SSQ-10 subset demonstrated weaker correlations with the depression benchmark, including stomach ache ($r_{pb}=0.190$), hallucinations ($r_{pb}=0.258$), and nightmares ($r_{pb}=0.293$), supporting the theoretical decision to refine the scale for depression specificity.

### Structural Validity and Stability

We conducted an Exploratory Factor Analysis (EFA) to determine the optimal factor structure, comparing the performance of the original SSQ-14, the existing SSQ-08, and the theoretical SSQ-10. ### The Existing Structure (SSQ-14 and SSQ-08)

For the established versions of the instrument, we observed a distinct methodological divergence between retention criteria. For the original SSQ-14 (<a href="#fig-factor-retention" class="quarto-xref">Figure 1</a>, Panel A), the Kaiser criterion ([Kaiser 1974](#ref-kaiser1974index)) favored a 2-factor solution in 87.5% of iterations, whereas Parallel Analysis (PA) ([Hayton, Allen, and Scarpello 2004](#ref-hayton2004factor)) suggested a more complex 3-factor structure (probability = 0.696).

Similarly, for the brief SSQ-08 (<a href="#fig-factor-retention" class="quarto-xref">Figure 1</a>, Panel C), the Kaiser criterion was unequivocal, suggesting a unidimensional (1-factor) structure in 100% of iterations. In contrast, PA continued to suggest multidimensionality, retaining 3 factors in 66.3% of the bootstrapped samples. Examination of the item clustering stability in Supplementary (Supp.) <a href="#fig-figure1_supplementary" class="quarto-xref">Figure 3</a> provides granular insight into these structural divergences. For the SSQ-14 (Supp. <a href="#fig-figure1_supplementary" class="quarto-xref">Figure 3</a>, Panel A), the 2-factor solution favored by the Kaiser criterion—which emerged as the consensus model with a stability of r cm_ka_ssq14—demonstrates high internal stability, with items consistently partitioning into distinct cognitive-somatic and affective subdomains. Most notably, thinking too much (SSQ01), difficulty concentrating (SSQ02), and irritability (SSQ03) co-occurred with greater than 95% probability across both extraction methods, identifying them as a robust core symptom cluster within the original tool.

However, the 3-factor solution suggested by Parallel Analysis (Supp. <a href="#fig-figure1_supplementary" class="quarto-xref">Figure 3</a>, Panel B), which achieved a consensus stability of r cm_pa_ssq14, reveals that the added dimensionality in the full scale is largely driven by the distinct clustering of somatic and perceptual items (e.g., hallucinations, nightmares) into a separate factor. Critically, while the existing SSQ-08 subset (Supp. <a href="#fig-figure1_supplementary" class="quarto-xref">Figure 3</a>, Panel C) retains thinking too much (SSQ01), it excludes the highly stable partners of this core cluster, specifically SSQ02 and SSQ03. The omission of these clinically coherent neurocognitive and affective markers, despite their strong structural cohesion with the primary idiom of distress (kufungisisa), suggests a potential limitation in capturing the full breadth of depressive symptomatology in this specific adolescent cohort. In this cohort, Parallel Analysis of the SSQ-08 revealed factors largely splitting between 2-, 3-, and 4-factor solutions, yet yielded a winning consensus model with a stability of 1713/6629 (25.8%); 95% CI: (24.8%, 26.9%). It is important to note, however, that the SSQ-08 was designed and validated to screen for broader Common Mental Disorders (CMD) in populations with high HIV prevalence, rather than depression exclusively. This finding explicitly validates the configuration of the SSQ-10, which retains this entire stable core cluster while systematically removing only the unstable ‘noise’ items (e.g., SSQ04–07) that drove the artifactual dimensionality in the original scale.

### The Theoretical Subset Divergence (SSQ-10)

The theoretical SSQ-10 subset (<a href="#fig-factor-retention" class="quarto-xref">Figure 1</a>, Panel B) exhibited a similar but pronounced conflict between methods. The Kaiser criterion strongly supported a unidimensional structure, retaining 1 factor in 83.1% of iterations ($n=8311$). Conversely, PA suggested a 3-factor solution in 72.2% of iterations ($n=7218$). This robust statistical disagreement—with one method suggesting a single score and the other suggesting three distinct sub-scales—necessitated a deeper investigation to determine whether the third factor represented a substantive latent construct or a statistical artifact.

### Artifact Investigation and Resolution

To adjudicate the conflicting retention suggestions for the SSQ-10, we examined the clustering of items within the 3-factor solution suggested by Parallel Analysis (consensus stability: 2891/7218 (40.1%); 95% CI: (38.9%, 41.2%)). As illustrated in <a href="#fig-heatmap-ssq10-pa" class="quarto-xref">Figure 2</a> (Panel A), the factor loading pattern revealed a distinct anomaly: while two stable factors emerged representing cognitive and affective symptoms (Factors 1 and 2), the third dimension was revealed to be a “singleton” factor comprised solely of poor sleep (SSQ08). This item was separated from both the cognitive (SSQ01–SSQ03) and general affective (SSQ09–SSQ14) clusters. However, closer inspection of the co-occurrence probabilities reveals that poor sleep (SSQ08) co-occurs significantly more often with the affective items (Factor 1: SSQ09–SSQ14) than with the cognitive items (Factor 2: SSQ01–SSQ03). Specifically, Factor 3 exhibited an average co-occurrence of 29.4% with Factor 1 items compared to only 0.3% with Factor 2 items. This pattern suggests that while statistically isolated in the 3-factor solution, the item shares more residual variance with the affective domain, and the isolation is likely driven by item-specific variance or “noise” rather than a substantive multidimensional trait.

Given that a single-item factor lacks theoretical parsimony, we evaluated the stability of the 2-factor solution as the next most plausible structure. Although the 2-factor solution was suggested as the primary structure in only 9% ($899/10,000$) of the initial Parallel Analysis iterations, deeper investigation revealed it to be highly robust when statistical artifacts were resolved. Specifically, the consensus stability for this 2-factor configuration was 846/899 (94.1%); 95% CI: (92.6%, 95.6%). This finding confirms that once the artifactual singleton factor is discounted, the SSQ-10 stabilizes into two robust dimensions: a cognitive-functional factor and an affective-emotional factor.
