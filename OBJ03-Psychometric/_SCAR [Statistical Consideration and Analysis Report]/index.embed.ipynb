## Introduction

The primary aim of this study is to establish a rigorous, evidence-based diagnostic cut-off score for the Shona Symptom Questionnaire (SSQ), specifically the theoretically derived 10-item subset (SSQ-10), using the Patient Health Questionnaire-9 (PHQ-9) as a reference standard for probable depression. To achieve this aim, the study first evaluates the structural validity and stability of the SSQ by comparing the original 14-item scale against the theoretical SSQ-10 and the existing 8-item Dixon-Chibanda subset \[@chibanda2019ssq8\] to identify the optimal configuration for clinical use. Subsequently, the analysis validates measurement invariance to confirm that the SSQ-10 functions consistently across depression status, biological sex, and age groups, ensuring the derived cut-off is universally applicable without bias \[@putnick2016measurement\]. Finally, the study derives a harmonized diagnostic threshold to facilitate accurate depression screening and comparability between the SSQ-10 and the PHQ-9 among adolescents and young adults.

## Methods

**Setting and Study Design**

This study is embedded within the Isisekelo Sempilo (“Foundation of Life”) trial, a $2 \times 2$ factorial randomized controlled trial conducted by the Africa Health Research Institute (AHRI) in the uMkhanyakude district of rural KwaZulu-Natal, South Africa \[@chidumwa2022isisekelo\]. The trial aimed to evaluate interventions integrating HIV prevention and care within sexual and reproductive health services. The study setting is predominantly rural and characterized by high levels of poverty and unemployment, with over 85% of young adults aged 20–24 reported as unemployed. The AHRI demographic surveillance area covers a population of approximately 140,000, characterized by a high prevalence of HIV.

**Participants and Sampling**

The trial utilized the AHRI demographic surveillance system as a sampling frame to randomly select 3,000 men and women aged 16–29 years, stratified by sex. Eligibility criteria for the main trial included residence within the surveillance area, willingness to provide informed consent, and willingness to participate in follow-up assessments including HIV testing via dried blood spot.

For the present psychometric analysis, a specific subsample was selected based on age, restricting the cohort to participants aged 16–24 years, and data completeness, requiring concurrent baseline measurements for both the SSQ-14 and the PHQ-9. This selection ensures that the validation and harmonization analyses are conducted on a directly comparable cohort relevant to adolescent and young adult mental health.

**Data Collection Procedures**

Participants were invited to participate via home visits conducted by trained fieldworkers. Upon contact, the study purpose was explained, and informed consent was obtained (written consent for adults; assent plus parental/guardian consent for participants aged 16–17).

Participants completed a structured baseline questionnaire administered electronically by fieldworkers using secure tablets. The interview included sociodemographic details, health care utilization history, and the full SSQ-14 and PHQ-9 instruments. To minimize social desirability bias and ensure comprehension, interviews were conducted in the local language, isiZulu, in private settings within the participant’s household or a preferred community location. Clinical attendance data were captured electronically via the ClinicLink system across 11 clinics within the surveillance area.

**Measures**

*The Patient Health Questionnaire (PHQ-9)*

The PHQ-9 is the depression module of the PRIME-MD diagnostic instrument, designed to align strictly with the criteria for Major Depressive Disorder (MDD) in the DSM-IV \[@Kroenke2001\]. It consists of 9 items scored on a 4-point Likert scale ranging from “Not at all” (0) to “Nearly every day” (3). It measures the severity of MDD symptoms over the past 2 weeks, covering domains such as anhedonia, depressed mood, sleep disturbance, fatigue, and suicidal ideation. While validated in South Africa, its etic approach may miss non-DSM manifestations of distress common in African contexts.

*The Shona Symptom Questionnaire (SSQ-14)*

The SSQ-14 represents an emic alternative. Developed in Zimbabwe, it was derived from ethnographic research into how Shona speakers articulate mental distress \[@Patel1997\]. It consists of 14 items with a binary (Yes/No) response format and measures Common Mental Disorders (CMD)—a blend of depression and anxiety—over the past 7 days. It includes indigenous idioms such as kufungisisa (“thinking too much”) and somatic sensations. While culturally resonant, its binary format and mixing of anxiety and depression items present specific challenges for harmonisation with the polytomous, depression-specific PHQ-9 \[@Haney2014\].

*Derivation of the SSQ-10 (Theoretical Subset)*

To enhance the specificity of the SSQ for depression screening, a theoretically derived subset (SSQ-10) was configured by retaining items that conceptually align with DSM-5 and ICD-11 criteria for a depressive episode, while excluding non-specific distress items \[@Patel1997\].

Ten items were retained: thinking too much (cognitive distress/rumination), difficulty concentrating (cognitive), irritability (affective-arousal), poor sleep (somatic-affective), crying (affective), feeling tired (somatic-affective), suicidal thoughts (cognitive-affective), general unhappiness (affective-cognitive), work lagging behind (functional/cognitive), and difficulty deciding (cognitive). Irritability was specifically retained as it is recognized as an alternative mood expression for youth depression in DSM-5.

Four items were excluded from the theoretical subset due to conceptual divergence from the core depression construct. Nightmares (SSQ04) were removed as they overlap with sleep disturbance but align more closely with anxiety or trauma. Seeing or hearing things (SSQ05) was excluded as it represents perceptual disturbances associated with severe or psychotic depression rather than core depressive criteria. Stomach ache (SSQ06) was excluded to reduce confounding with general somatic complaints common in this setting, while being frightened by trivial things (SSQ07) was removed as it serves as a marker for anxiety or hyperarousal rather than depression.

**Data Analysis**

All statistical analyses were conducted using R Statistical Software \[@rcoreteam2023\].

*Descriptive Statistics*

Descriptive statistics were computed to summarize the demographic and clinical characteristics of the sample. Categorical variables (e.g., biological sex, employment status, depression status) were reported as frequencies and percentages. Continuous variables (e.g., age, PHQ-9 total score) were summarized using means and standard deviations (SD). The distribution of item responses was examined to identify endorsement rates and potential floor or ceiling effects.

*Structural Validity and Stability*

To determine the most robust version of the SSQ for screening purposes, three item configurations were evaluated: the original 14-item scale (SSQ-14), a theoretically derived 10-item subset focusing on core depressive symptoms (SSQ-10), and the 8-item Dixon-Chibanda subset \[@chibanda2019ssq8\].

A Monte Carlo split-sample resampling approach was employed to rigorously test the stability of these structures. The dataset was randomly partitioned into exploratory and confirmatory halves for 10,000 iterations. In each iteration, Exploratory Factor Analysis (EFA) was conducted to identify the most likely factor structure using established criteria appropriate for categorical data, specifically the Kaiser Criterion and Parallel Analysis, utilizing polychoric correlations \[@holgado2010polychoric\]. The Consensus Model was then identified as the specific grouping of items that appeared most frequently across these replications, serving as the definitive structure for each scale version.

Following the exploratory phase, the stability of the Consensus Model for the SSQ-10 was tested using Confirmatory Factor Analysis (CFA) on the hold-out data samples using the Weighted Least Squares Mean and Variance adjusted (WLSMV) estimator.

*Reliability and Measurement Invariance*

Internal consistency reliability was calculated using both Cronbach’s Alpha and McDonald’s Omega Total to provide a comprehensive assessment of the scale’s precision.

To ensure the scale performs fairly across different populations, measurement invariance was assessed. This procedure evaluated whether the SSQ-10 measures depression consistently regardless of a participant’s biological sex, age group, or depression severity status \[@putnick2016measurement\].

*Diagnostic Cut-offs and Harmonisation*

To drive diagnostic cut-offs and harmonisation, an optimal screening cut-off for the SSQ-10 was established using the PHQ-9 as the clinical reference standard. Two complementary methods were applied. First, Receiver Operating Characteristic (ROC) analysis was used to identify the score threshold that maximized the balance between sensitivity and specificity. Second, Item Response Theory (IRT) was employed to overcome the limitations of simple binary cut-offs by linking the SSQ-10 and PHQ-9 onto a common latent metric of depression severity \[@samejima1969estimation\]. This allowed for the identification of the precise level of underlying depression corresponding to the PHQ-9 threshold and mapping it directly to the SSQ-10, creating a harmonized standard for classification \[@kolen2014test\].

**The Theoretical Subset Divergence (SSQ-10)**

The theoretical SSQ-10 subset (@fig-factor-retention, Panel C) exhibited a similar but pronounced conflict between methods. The Kaiser criterion strongly supported a unidimensional structure, retaining 1 factor in 83.1% of iterations ($n=8311$). Conversely, PA suggested a 3-factor solution in 72.2% of iterations ($n=7218$). This robust statistical disagreement—with one method suggesting a single score and the other suggesting three distinct sub-scales—necessitated a deeper investigation to determine whether the third factor represented a substantive latent construct or a statistical artifact.

**Artifact Investigation and Resolution**

To adjudicate the conflicting retention suggestions for the SSQ-10, we examined the clustering of items within the 3-factor solution suggested by Parallel Analysis (consensus stability: 2891/7218 (40.1%); 95% CI: (38.9%, 41.2%)). As illustrated in @fig-heatmap-ssq10-pa (Panel A), the factor loading pattern revealed a distinct anomaly: while two stable factors emerged representing cognitive and affective symptoms (Factors 1 and 2), the third dimension was revealed to be a “singleton” factor comprised solely of poor sleep (SSQ08). This item was separated from both the cognitive (SSQ01–SSQ03) and general affective (SSQ09–SSQ14) clusters. However, closer inspection of the co-occurrence probabilities reveals that poor sleep (SSQ08) co-occurs significantly more often with the affective items (Factor 1: SSQ09–SSQ14) than with the cognitive items (Factor 2: SSQ01–SSQ03). Specifically, Factor 3 exhibited an average co-occurrence of 29.4% with Factor 1 items compared to only 0.3% with Factor 2 items. This pattern suggests that while statistically isolated in the 3-factor solution, the item shares more residual variance with the affective domain, and the isolation is likely driven by item-specific variance or “noise” rather than a substantive multidimensional trait. Given that a single-item factor lacks theoretical parsimony, we evaluated the stability of the 2-factor solution as the next most plausible structure. Although the 2-factor solution was suggested as the primary structure in only 9% ($899/10,000$) of the initial Parallel Analysis iterations, deeper investigation revealed it to be highly robust when statistical artifacts were resolved. Specifically, the consensus stability for this 2-factor configuration was 846/899 (94.1%); 95% CI: (92.6%, 95.6%). This finding confirms that once the artifactual singleton factor is discounted, the SSQ-10 stabilizes into two robust dimensions: a cognitive-functional factor and an affective-emotional factor.

**Confirmatory Validation and Model Comparison**

Following the discovery phase, we subjected the identified structures to rigorous testing using Confirmatory Factor Analysis (CFA) with the Weighted Least Squares Mean and Variance adjusted (WLSMV) estimator.

*Model Fit and Scoring Justification*

We evaluated the stability and fit of the 1-Factor and 2-Factor models for both the SSQ-10 and the PHQ-9 reference standard across 10,000 Monte Carlo samples. @tbl-model-comp presents the global fit indices, internal consistency reliability, and latent factor correlations.

For the PHQ-9, while the 1-Factor model showed excellent CFI (0.994), the RMSEA was marginal (0.076), with only 64.8% of samples meeting the $\le$ 0.08 threshold. The 2-Factor model improved the RMSEA to 0.049. However, the high correlation between the two factors ($\phi = 0.872$) and the overlapping reliability estimates ($\alpha$ & $\omega$ \> .90 for both total and subscales, with 95% CIs consistently excluding values below .90) support the retention of the unidimensional scoring convention.

For the SSQ-10, the contrast was more pronounced. The 1-Factor model demonstrated a poor RMSEA of 0.091 (95% CI: 0.075, 0.107), with only 7.6% of replications indicating acceptable fit. In contrast, the 2-Factor consensus model achieved a near-perfect fit (CFI = 1.000; RMSEA = 0.010), with 100% of samples meeting adequacy thresholds. Despite this structural distinctness, the moderate-to-strong factor correlation ($\phi = 0.792$) and high internal consistency for both the global scale ($\alpha$ & $\omega$ \> .90; 95% CIs \> .96) and subscales (all 95% CI lower bounds \> .93) suggest that the total score remains a meaningful metric for general depression screening, capturing a cohesive underlying construct despite the multidimensional nuances.

```
# Helper to build CFA stability tables for a 1-factor vs 2-factor solution
build_cfa_stability_table <- function(results_1factor, results_2factor) {
  
  # Fit indices -------------------------------------------------------------
  fit_tbl <-
    dplyr::bind_rows(
      results_1factor[["gt_tables"]][["gt_fit_source"]] |>
        dplyr::mutate(factor = "1-Factor"),
      results_2factor[["gt_tables"]][["gt_fit_source"]] |>
        dplyr::mutate(factor = "2-Factor")
    ) |>
    dplyr::mutate(
      Prop_Good_CFI   = sprintf("%1.1f", Prop_Good_CFI * 100),
      Prop_Good_RMSEA = sprintf("%1.1f", Prop_Good_RMSEA * 100)
    ) |>
    dplyr::select(
      factor,
      `CFI (95% CI)`, Prop_Good_CFI,
      `RMSEA (95% CI)`, Prop_Good_RMSEA
    )
  
  # Factor correlations (only for 2-factor solution) -----------------------
  factor_corr_tbl <-
    results_2factor[["gt_tables"]][["gt_factor_correlations_source"]] |>
    dplyr::transmute(
      `Factors Corr (95% CI)` = `Corr (95% CI)`,
      factor = "2-Factor"
    )
  
  # Reliability: collapse scales per solution ------------------------------
  reliability_tbl <-
    dplyr::bind_rows(
      results_2factor[["gt_tables"]][["gt_reliability_source"]] |>
        dplyr::mutate(factor = "2-Factor"),
      results_1factor[["gt_tables"]][["gt_reliability_source"]] |>
        dplyr::mutate(factor = "1-Factor")
    ) |>
    dplyr::select(factor, Scale, `Alpha (95% CI)`, `Omega (95% CI)`) |>
    dplyr::group_by(factor) |>
    dplyr::summarise(
      `Alpha (95% CI)` = paste0(Scale, ": ", `Alpha (95% CI)`, collapse = "\n"),
      `Omega (95% CI)` = paste0(Scale, ": ", `Omega (95% CI)`, collapse = "\n"),
      .groups = "drop"
    )
  
  # Combine, reshape, and create row groups --------------------------------
  combined <- fit_tbl |>
    dplyr::left_join(factor_corr_tbl, by = "factor") |>
    dplyr::left_join(reliability_tbl, by = "factor") |>
    dplyr::rename(
      Solution         = factor,
      `% CFI ≥ 0.90`   = Prop_Good_CFI,
      `% RMSEA ≤ 0.08` = Prop_Good_RMSEA
    )
  
  combined_long <- combined |>
    tidyr::pivot_longer(
      cols      = -Solution,
      names_to  = "Measure",
      values_to = "Statistic"
    ) |>
    dplyr::mutate(
      Group = dplyr::case_when(
        Measure %in% c("CFI (95% CI)", "% CFI ≥ 0.90",
                       "RMSEA (95% CI)", "% RMSEA ≤ 0.08") ~ "Fit indices",
        Measure %in% c("Alpha (95% CI)", "Omega (95% CI)") ~ "Reliability",
        Measure %in% c("Factors Corr (95% CI)") ~ "Factor correlations",
        TRUE ~ "Other"
      ),
      Measure = factor(
        Measure,
        levels = c(
          "CFI (95% CI)", "% CFI ≥ 0.90",
          "RMSEA (95% CI)", "% RMSEA ≤ 0.08",
          "Alpha (95% CI)", "Omega (95% CI)",
          "Factors Corr (95% CI)"
        )
      ),
      Group = factor(
        Group,
        levels = c("Fit indices", "Reliability", "Factor correlations", "Other")
      )
    ) |>
    dplyr::arrange(Group, Measure)
  
  combined_long |>
    tidyr::pivot_wider(
      names_from  = Solution,
      values_from = Statistic
    ) |>
    dplyr::mutate(
      Measure = as.character(Measure),
      Group   = as.character(Group)
    ) |>
    dplyr::select(Group, Measure, dplyr::everything())
}

# Build tables for PHQ-9 and SSQ-10 ----------------------------------------

table_phq09_cfa_stability <- build_cfa_stability_table(
  results_1factor = batch_res$phq09_data_driven,
  results_2factor = batch_res$phq09_pa_3f
)

# Remove first columns (Group/Measure) from SSQ-10 table before binding ----

table_ssq10_cfa_stability <- build_cfa_stability_table(
  results_1factor = batch_res$ssq10_data_driven,
  results_2factor = batch_res$ssq10_pa_3f
) |>
  dplyr::select(-Group, -Measure)

# Clean factor column names and prepare for column groups -------------------

table_phq09_cfa_stability <- table_phq09_cfa_stability |>
  dplyr::rename(
    phq9_1factor = `1-Factor`,
    phq9_2factor = `2-Factor`
  )

table_ssq10_cfa_stability <- table_ssq10_cfa_stability |>
  dplyr::rename(
    ssq10_1factor = `1-Factor`,
    ssq10_2factor = `2-Factor`
  )

combined_tbl <- table_phq09_cfa_stability |>
  dplyr::bind_cols(table_ssq10_cfa_stability)

# Final GT table with column and row groups --------------------------------

combined_tbl |>
  gt::gt(groupname_col = "Group") |>
  gt::sub_missing(columns = dplyr::everything(), missing_text = "-") |>
  # Center all non-group columns
  gt::cols_align(
    align   = "center",
    columns = -c(Group, Measure)
  ) |>
  # Clean column labels
  gt::cols_label(
    Measure       = "Measure",
    phq9_1factor  = "1-Factor",
    phq9_2factor  = "2-Factor",
    ssq10_1factor = "1-Factor",
    ssq10_2factor = "2-Factor"
  ) |>
  # Column groups (spanners)
  gt::tab_spanner(
    label   = "PHQ-9",
    columns = c(phq9_1factor, phq9_2factor)
  ) |>
  gt::tab_spanner(
    label   = "SSQ-10",
    columns = c(ssq10_1factor, ssq10_2factor)
  ) |>
  # Bold row groups and column groups
  gt::tab_style(
    style = gt::cell_text(weight = "bold"),
    locations = gt::cells_row_groups()
  ) |>
  gt::tab_style(
    style = gt::cell_text(weight = "bold", align = "center"),
    locations = gt::cells_column_spanners()
  ) |>
  gt::tab_header(
    title    = "Model Comparison (1-Factor vs 2-Factor)",
    subtitle = paste(
      c(
        "Across Monte Carlo Samples: Global Fit Indices",
        "Latent Factor Correlations",
        "Internal Consistency Reliability"
      ),
      collapse = "; "
    )
  ) |>
  gt::tab_source_note(
    source_note = paste(
      c(
        "Note:",
        "CFI/TLI ≥ 0.90 (often ≥ 0.95 for good fit) and RMSEA ≤ 0.08 (≤ 0.06 for good fit) indicate acceptable model fit.",
        "Percentage correspond to the fraction of Monte Carlo samples meeting these benchmarks.",
        "Latent factor correlations estimated from CFA models across Monte Carlo splits, reported as mean and 95% Monte Carlo quantile interval.",
        "Interpretation guide: α, ω ≥ 0.70 acceptable, ≥ 0.80 good, ≥ 0.90 excellent.",
        "Estimates are averaged across Monte Carlo splits with 95% quantile intervals."
      ),
      collapse = "\n"
    )
  )
```

*Measurement Invariance*

To ensure the equitable utility of the SSQ-10 across different population subgroups, we conducted Multi-Group CFA (MGCFA) using 1,000 Monte Carlo resampling iterations. We tested for measurement invariance across biological sex (Male vs. Female) and age groups (16–19 vs. 20–24). The results for the SSQ-10 (1-Factor & 2-Factor) and PHQ-9 (1-Factor & 2-Factor) models are summarized in Supplementary Table S1.

For the PHQ-9 1-Factor model, the analysis demonstrated robust invariance based on the Comparative Fit Index. Across age groups, 99.7% of Monte Carlo samples met the change in CFI criterion ($\le$ 0.010), and across biological sex, 99.5% met this threshold. However, the Root Mean Square Error of Approximation proved highly unstable, with only 16.8% of age group samples and 44.4% of sex samples meeting the invariance criterion ($\le$ 0.015).

Similarly, the PHQ-9 2-Factor model showed perfect stability regarding the Comparative Fit Index, with 100% of samples meeting the criterion across both age and sex groupings. Yet, the Root Mean Square Error of Approximation remained inconsistent, satisfying the threshold in only 20.3% of age group samples and 46.4% of sex samples. Despite the divergence in the error of approximation metric, the high stability of the Comparative Fit Index supports the conclusion that the PHQ-9 constructs are measured comparably across these groups, as the Comparative Fit Index is often considered a more reliable indicator of invariance in this context.

For the SSQ-10 1-Factor Model, the analysis provided strong evidence for scalar invariance. Across Age Groups, the mean $\Delta$CFI was 0.000, with 96.7% of samples meeting the invariance threshold, and 87.5% meeting the RMSEA criterion. Similarly, across Biological Sex, 95.2% of samples satisfied the $\Delta$CFI cutoff, and 94.9% met the $\Delta$RMSEA criterion. These results indicate that the unidimensional structure of the SSQ-10 is measurement invariant across these demographic factors.

For the SSQ-10 2-Factor Model, the analysis also provided strong support for scalar invariance across Biological Sex. The mean change in CFI was 0.000, with 99.8% of Monte Carlo samples meeting the strict invariance threshold. Root Mean Square Error of Approximation stability was also high, with 89.5% of samples meeting the criterion. For Age Groups, the Comparative Fit Index criterion remained highly robust, with 99.7% of samples meeting the $\le$ 0.010 threshold. However, the Root Mean Square Error of Approximation metric was less stable, with only 56.4% of samples meeting the strict $\le$ 0.015 cutoff. This divergence likely reflects the known sensitivity of the Root Mean Square Error of Approximation to sample size and model complexity in subgroup partitions. Given the near-perfect stability of the Comparative Fit Index, these findings suggest that the SSQ-10 measures the underlying construct of depression in a comparable manner across these key demographic groups, satisfying the prerequisites for valid group comparisons \[@putnick2016measurement\].
