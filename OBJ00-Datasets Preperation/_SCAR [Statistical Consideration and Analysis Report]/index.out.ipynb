# Project Overview & Study Design

The CO-LUMINATE dataset links individual-level mental health outcomes from specific clinical cohorts to longitudinal household data from the AHRI Health and Demographic Surveillance System (HDSS).

## Core Design

### Exposure Window (Developmental Origins)

-   **Early Childhood:** Ages 0–5 years  
-   **Middle Childhood:** Ages 6–10 years  
-   **Source:** AHRI HDSS (annual panels used to derive latent trajectory classes)

### Outcome Window (Adolescence/Young Adulthood)

-   **Observation Period:** Ages 13–24 years  
-   **Source:** 5 nested clinical studies (Depression & Behavior)

## The “Hub-and-Spoke” Linkage

The dataset relies on linking the **USUBJID** from the clinical visits (Spokes) to the historical residency timeline (Hub) in the HDSS.

## Data Sources (The 5 Cohorts)

The outcome data is derived from five distinct studies. The `STUDY` variable is coded as a factor with the following fixed order:

| Factor Level | Study Name | N (Approx) | Primary MH Tool | Role in Analysis |
|-----------|--------------|---------:|-------------|--------------------------|
| 1 | Dreams | \[N\] | SSQ-14 | Outcome (Needs Calibration) |
| 2 | Isisekelo Sempilo | \[N\] | PHQ-9 & SSQ-14 | Calibration Bridge / Gold Standard |
| 3 | Multilevel | \[N\] | SSQ-14 | Outcome (Needs Calibration) |
| 4 | TasP | \[N\] | PHQ-9 | Outcome (Gold Standard) |
| 5 | Thetha Nami | \[N\] | PHQ-9 | Outcome (Gold Standard) |

# CO-LUMINATE Secondary Data Analysis

The data-driven aspect of this project entails the creation of the CO-LUMINATE longitudinal linked dataset. This dataset will allow us to understand the longitudinal relationships between socioeconomic status (SES), food security (captured through HDSS-based household indicators), caregiver residency and co-residency, and depressed mood (measured at multiple time points), among rural South African young people aged 13–24. These findings will be used to generate and test hypotheses regarding the mechanisms underpinning young people’s depression in low- and middle-income countries (LMICs).

The CO-LUMINATE longitudinal dataset will link existing data from 5,279 young people aged 13–24 years, for whom we have rigorous measures of depressed mood at one or more time points from AHRI’s past research studies, with their households’ HDSS data. Through this process, each young person’s depression data is linked with an extensive longitudinal database containing information on their household’s SES and food security–related conditions, as well as household composition. The HDSS also allows us to link young people to their caregivers, so that we can examine patterns of co-residence between a young person’s caregivers, and between them and their caregivers.

The novel contribution of this project, and our ability to do this innovative work, arises from three facts:

1.  AHRI’s HDSS, like most, does not include rigorous individual-level measurement of young people’s mental health outcomes, but offers an opportunity to examine social exposures longitudinally.
2.  Over the past decade, AHRI has conducted cohort studies among young people which included rigorous measurement of depressed mood (SSQ-14 and PHQ-9) at one or more time points, providing a unique data resource.
3.  By bringing these two types of data together, we are able to examine – in a sample of 5,279 – patterns of exposure to social determinants and trajectories of young people’s mental health (with a focus on depression).

The cohort studies used for this project – DREAMS, Multilevel, Isisekelo Sempilo, Treatment as Prevention (TasP), and Thetha Nami – were nested within the HDSS. This allows us to link young people’s individual-level depression data gathered in these studies with their household’s HDSS data. This linkage creates an unprecedented longitudinal dataset of household social exposures and depression in young people.

The basis for this project has already begun. In partnership with Aga Khan University, KEMRI, and the University of Michigan Center for Global Health Equity, co-applicant NM integrated data from the nested studies into a harmonised individual-level mental health dataset for pooled analyses. The harmonised individual-level mental health dataset can now be linked to the study participants’ household HDSS data. This allows longitudinal measures of young people’s social and environmental exposures at the household level to be explored in relation to their depression symptoms.

# Study Setting and Design

## Study Setting

AHRI is a transdisciplinary scientific research institute in KwaZulu-Natal, South Africa. AHRI’s Somkhele campus is located in an under-resourced rural area centred in a Health and Demographic Surveillance System (HDSS). The HDSS conducts tri-annual surveys of approximately 22,000 households and annual surveys of roughly 150,000 individuals in an area of 845 km². The area typifies the challenges that face young people in LMICs: poverty, food insecurity, high HIV prevalence, migration, and substance abuse, all increasing vulnerability to depression.

# Methodological Algorithms

## The “Age-Splitter” Algorithm (Data Preparation)

To prepare the input data for trajectory modelling:

### Episode Expansion

AHRI residency episodes are intersected with the participant’s specific “Age 0–10” timeline.

### Annualization

Continuous residency episodes are sliced into 1-year intervals (Age 0, Age 1, etc.).

### Time-Weighting

Annual exposure values are calculated based on the household state where the child spent the majority of that specific age-year.

## Trajectory & Mixture Modelling (Derivation)

Exposure classes are derived using Multi-Trajectory Modeling (MTM) or Joint Latent Class Analysis to identify subgroups of individuals following similar patterns across all base variables simultaneously.

### Continuous Exposures (Child Count)

Growth Mixture Modelling (GMM) is used to identify latent classes with distinct intercepts and slopes.

### Categorical/Ordinal Exposures (SES, Parental Absence, Education)

Latent Class Analysis (LCA) or Latent Transition Analysis (LTA) is used to identify subgroups based on patterns of endorsement across the 10 time points.

### Strategy 1: Global (0–10 Years)

Models are fitted using the full 11-year sequence of all base variables. This creates a single **Childhood Profile** (`Global_Adversity_Class`) for every participant.

### Strategy 2: Sensitive Periods (Split)

Models are fitted separately for:

-   **0–5 subset:** `Early_Adversity_Class`  
-   **6–10 subset:** `Mid_Adversity_Class`

This creates two distinct exposure variables to test sensitive period hypotheses.

# Psychometric Harmonization (Triangulation Strategy)

To ensure robustness, the project maintains two parallel definitions of “Depression” derived from the Isisekelo Sempilo bridge sample.

## Method 1: Traditional Binary (ROC-Based)

-   **Variable:** `Depression_Bin_ROC`  
-   **Approach:** Standard clinical epidemiology.  
-   **Process:**
    -   Identify the optimal integer cut-off on the raw SSQ-14 scale (e.g., ≥ 8) that maximizes sensitivity and specificity relative to the PHQ-9.  
-   **Use Case:**
    -   Comparison with previous literature and simpler prevalence reporting.

## Method 2: Latent Harmonization (IRT-Based)

-   **Variables:** `Harmonized_Theta` (Continuous) & `Depression_Bin_IRT` (Binary).  
-   **Approach:** Psychometric measurement invariance.  
-   **Process:**
    -   Fit a Graded Response Model to estimate a common latent trait ($\theta$).  
    -   Identify the $\theta$ value corresponding to PHQ-9 = 10.  
    -   Classify all participants based on this latent threshold.
-   **Use Case:**
    -   Primary analytical models (Regression/Mediation) where measurement precision and invariance are critical.

# Quality Control & Missingness Codes

## Missing Data Flags

The following codes are standardized across data frames:

-   `NA`: System missing / Not applicable (e.g., PHQ-9 score for a Dreams participant).  
-   `-99`: Refused to answer.  
-   `-88`: Don’t know.  
-   `-77`: Not located / Lost to follow-up.

## Exclusion Criteria (Consort Flow)

Participants are excluded if:

-   They do not have a valid `USUBJID` linking to the HDSS Masterfile.  
-   They have \< 180 days of residency observation during the Exposure Window (0–10 years).  
-   They have missing or invalid Date of Birth.
